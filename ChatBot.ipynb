# Chatbot

This chat bot gives details about indian cars in a small Automative Showroom.

```
def greet_and_introduce():
    responses = "Hi I am chatbot. I can help you to know about car and its models"
    print(responses)

def show_menu():
    print("These are the companies")
    print("1.Mahindra" )
    print("2.Tata motors" )
    print("3.Renault" )
    print("4.Ford India")
    return int(input("enter your choice [1-4]: "))

def mahindra():
    print("These are the models")
    print('5.Mahindra Bolero')
    print('6.Mahindra thar')
    print('7.Mahindra kuv nxt')
    return int(input("enter your choice[5-7]: "))
    

def mahindra_bolero():
    print("Name : Mahindra Bolero")
    print("price : 8-9 lakhs")
    print("year: 2020 ")
    return("colour:black,white")
    
def mahindra_thar():
    print("Name : Mahindra thar")
    print("price:9.8 lakh")
    print("year: 2019 ")
    return("colour:red,black,white")

def mahindra_kuvnxt():
    print("Name : mahindra kuv nxt")
    print("price:5-6 lakh")
    print("year:2019")
    return("colour:red,black,white,orange")

def tatamotors():
    print("These are the models")
    print('8.Tata tiago')
    print('9.Tata tigor')
    print('10.Tata nexon')
    return int(input("enter your choice[8-10]: "))

def tata_tiago():
    print("Name:Tata tiago")
    print("price:4.7 to 5 lakh")
    print("year :2016")
    return("colour:yellow,red,black,white")

def tata_tigor():
    print("Name:Tata tigor")
    print("price:5 to 6 lakh")
    print("year:2017") 
    return("colour:red,black,white,blue")

def tata_nexon():
    print("Name:Tata nexon")
    print("price:6 to 12 lakh")
    print("year:2017")
    return("colour: green,silver,white")

def renault():
    print("These are the models")
    print("11.Renault Duster")
    print("12.Renault kwid")
    print("13.Renault triber")
    return int(input("enter your choice:[11-13]: "))

def renault_duster():
    print("Name : Renault Duster")
    print("price: 9 lakh")
    print("year: 2020")
    return("colour: blue,red,black,white")

def renault_kwid():
    print("Name : Renault kwid")
    print("price :  3 to 5 lakh")
    print("year : 2020")
    return("colour : ice blue,maroon")

def renault_triber():
    print("Name : Renault triber")
    print("price : 5 to 6 lakh")
    print("year : 2020")
    return("colour : yellow,red,white")

def fordindia():
    print("These are the models")
    print("14.Ford aspire")
    print("15.Ford freestyle")
    print("16.Ford figo")
    return int(input("enter your choice:[14-16]: "))

def ford_aspire():
    print("Name:Ford aspire")
    print("price:6.09 lakh")
    print("year:2018")
    return("colour:white,red,silver")

def ford_freestyle():
    print("Name:Ford freestyle")
    print("price:6 lakh")
    print("year:2019")
    return("colour:silver,white,red")

def ford_figo():
    print("Name:Ford figo")
    print("price: 5 lakh")
    print("year:2019")
    return("colour:white,red,yellow,black")

def bot():
    greet_and_introduce()
    choice = show_menu()
    #while choice != 3:
    if choice == 1:
        choice1=mahindra()
        if choice1==4:
            print(mahindra_bolero())
        elif choice1==5:
            print(mahindra_thar())    
        else:
            print(mahindra_kuvnxt())
    elif choice == 2:
        choice2=tatamotors()
        if choice2==7:
            print(tata_tiago)
        elif choice2==8:
            print(tata_tigor())
        else:
            print(tata_nexon())
    elif choice==3:
        choice3=renault()
        if choice3==10:
            print(renault_duster())
        elif choice3==11:
            print(renault_kwid())
        else:
            print(renault_triber())
    else:
        choice4=fordindia()
        if choice4==14:
            print(ford_aspire())
        elif choice==15:
            print(ford_freestyle())
        else:
            print(ford_figo())
bot()
```
